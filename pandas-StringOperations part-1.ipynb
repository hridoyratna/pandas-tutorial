```python
#problem in vectorized operations in vanilla python
s = ['cat','mat',None,'rat','c','cut']
[i.startswith('c') for i in s]   #Do not work for 'None' type object  and it comperatively slow for larger data
```

###common Functions

```python
#lower/ upper/ capitalize/ title
titanic['Name'].str.lower()
titanic['Name'].str.upper()
titanic['Name'].str.capitalize()
titanic['Name'].str.title()
```

```python
#len
titanic['Name'].str.len().sort_values(ascending=False)
```

```python
#Find the name which name length is maximum
titanic['Name'][titanic['Name'].str.len() == titanic['Name'].str.len().max()]
```

```python
#Sir answer
titanic['Name'][titanic['Name'].str.len() == 82]
```

```python
#strip  : Remove before and after irrelevent space
"           Hridoy         Ratno".strip()
"           Hridoy         ".strip()
"           Hridoy         Ratno            ".strip()
```

```python
titanic['Name'].str.strip()
```

```python
#split -> get
titanic['lastname'] = titanic['Name'].str.split(',').str.get(0)  #sarname
titanic['Name'].str.split(',').str.get(1)  #name
titanic.head()
```

```python
titanic[['title','firstname']] = titanic['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand= True)  #n means how many split we needed..1 refers to one split we need
#expand convert list to dataframe
titanic.head()
```

```python
titanic['title'].value_counts()
```

```python
#replace
titanic['title']=titanic['title'].str.replace('Ms.','Miss.')
titanic['title']=titanic['title'].str.replace('Mlle.','Miss.')
titanic['title'].value_counts()
```

###Filtering

```python
#startswith/endswith
titanic[titanic['firstname'].str.startswith('A')]
```

```python
#startswith/endswith
titanic[titanic['firstname'].str.endswith('A')]
```

```python
#isdigit/isalpha..
titanic[titanic['firstname'].str.isdigit()]
```

###Advance filtering

```python
#applying regex : contains
#search john -> both case
titanic[titanic['firstname'].str.contains('john',case=False)]
```

```python
#find lastname with start and end char vowel
titanic[titanic['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
#find lastname with start and end char consonant
titanic[titanic['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```

```python
#slicing
titanic['Name'].str[:4]   #Find first 4 char
```

```python
#Find alternate char of name
titanic['Name'].str[::2]   #jump 2 char
```
