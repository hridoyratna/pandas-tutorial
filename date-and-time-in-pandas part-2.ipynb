###Why separate objects to handle data and time when python already has datetimefuctionality?
<p>1. Syntex wise datetime is very covenient.</p>
<p>2. But the performance takes a hit while working with huge data. List vs Numpy Array</p>
<p>3. The weaknesses of python's datetime format inspired the Numpy team to add a set of native time series data type to numpy.</p>
<p>4. The datetime64 dtype encodes dates as 64-bit integers and thus allows arrays of dates to be represented very compactly.</p>

```python
import numpy as np
date = np.array('2015-07-04',dtype=np.datetime64)
date
```

```python
date + np.arange(12)
```

<p>1. Because of the uniform type in Numpy datetime64 arrays, this type of operation can be accomplished much more quickly than if we were working directly with python's datetime objects, especially as arrays get large.</p>
<p>2. Pandas Timestamp object combines the case-of-use of python datetime with the efficient storage and vectorized interface of numpy.daatetime64 </p>
<p>3. From group of these Timestamp objects, Pandas can construct a DatetimeIndex that can be used to index data in a Series or Dataframe</p>

###2. DatetimeIndex Object
<p>A collection of pandas timestamp</p>

```python
#from strings
pd.DatetimeIndex(['2023/1/1','2022/1/1','2024/2/2','2025/02/25'])
```

```python
type(pd.DatetimeIndex(['2023/1/1','2022/1/1','2024/2/2','2025/02/25']))
```

```python
pd.DatetimeIndex(['2023/1/1','2022/1/1','2024/2/2','2025/02/25'])[0]   #Timetamp object
```

```python
#using python datetime object
pd.DatetimeIndex([dt.datetime(2023,1,1),dt.datetime(2024,1,1),dt.datetime(2025,1,1)])
```

```python
#using pd.timestamps
pd.DatetimeIndex([pd.Timestamp(2023,1,1),pd.Timestamp(2024,1,1),pd.Timestamp(2025,1,1)])
```

```python
#using datetimeindex as series index
dt_index = pd.DatetimeIndex([pd.Timestamp(2023,1,1),pd.Timestamp(2024,1,1),pd.Timestamp(2025,1,1)])
pd.Series([1,2,3],index=dt_index)
```

###to_datetime function
<p>converts an existing objects to pandas timestamp/datetimeindex object</p>

```python
#simple series example
s = pd.Series(['2023/1/1','2022/1/1','2025/1/1','2026/1/1'])
s
```

```python
#If we find year above series ,we work hard .Like
s.str.split('/').str.get(0)
```

```python
#solution by convert object to integer
pd.to_datetime(s).dt.year
pd.to_datetime(s).dt.month
pd.to_datetime(s).dt.day
pd.to_datetime(s).dt.month_name()
pd.to_datetime(s).dt.day_name()
```

```python
#with errors
s = pd.Series(['2023/1/1','2022/1/1','2025/130/1','2026/1/1'])
pd.to_datetime(s)
```

```python
#solve above error
pd.to_datetime(s,errors='coerce')
pd.to_datetime(s,errors='coerce').dt.year
pd.to_datetime(s,errors='coerce').dt.month_name()
```
