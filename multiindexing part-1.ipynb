###Multiindexing Series
<p>Multiindex series also known as Hierarchical Indexing</p>
<p>Multiple index levels withina single Index</p>

```python
#Can we have multiple index? Let's try
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
a = pd.Series([1,2,3,4,5,6,7,8],index= index_val)
a
```

```python
#The problem?
a['cse']
```

```python
#The solution
a[('cse',2021)]
```

###How to create multiindex object
<p>1. pd.MultiIndex.from_tuples()</p>
<p>2. pd.MultiIndex.from_product()</p>

```python
#pd.MultiIndex.from_tuples()
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex.levels
multiindex.levels[0]
multiindex.levels[1]
```

```python
tuples = [(1, 'red'), (1, 'blue'),
          (2, 'red'), (2, 'blue')]
pd.MultiIndex.from_tuples(tuples, names=('number', 'color'))
```

```python
# pd.MultiIndex.from_product()
multiindex = pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])  #Cartesian product perfom when we pass 2D array
```

```python
#create a series with multiindex object
s = pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
s
```

```python
#how to fetch items from such a series
s['cse']
s[('ece',2020)]
```

```python
#a logical question to ask :How many dimention Above multiindex series
```

######Ans : 2D .Cause for finding elements we shoud have provided two elements .

```python
#Unstack: convert series to dataframe using insideal index as a column name
temp = s.unstack()
temp
```

```python
#Stack : convert dataframe to series using column name as a insideal index
temp.stack()
```

###Long Vs Wide Data
<p>Wide format is where we have a single rowww for every data point with multiple columns to hold the values of various attributes..</p>
<p>long format is where, for each data point wwe have as many rows the number of attributes andd each row contains the value of a particular attribute for a given data point..  </p>

```python
#melt -> simple example branch
# melt convert wide data to Long data
pd.DataFrame({'cse': [120]})
```

```python
pd.DataFrame({'cse': [120]}).melt()
```

```python
#melt -> branch with year
pd.DataFrame({'cse':[120],'ece' : [100],'mech':[50]})
```

```python
pd.DataFrame({'cse':[120],'ece' : [100],'mech':[50]}).melt(var_name='branch',value_name='num_students')
```

```python
temp=pd.DataFrame(
    {
       'branch' : ['cse','ece','mech'],
       '2020' : [100,150,60],
       '2021' : [120,130,80],
       '2022': [150,140,70],
    }
)
temp
```

```python
temp.melt()  # This is a problem here
```

```python
temp.melt(id_vars='branch',var_name='year',value_name='students')  #write way
```
