###Pandas Series

###Importing Pandas

```python
#Series
import numpy as np
import pandas as pd
```

###Different types are used to crate a series
<p>1.Using python list</p>
<p>2.Using python Dict</p>

###Series From list

```python
#String
country=['bd','In','Pk','china','USA','US']
pd.Series(country)
```

```python
#integers
runs=[13,54,100,58,90]
run_series=pd.Series(runs)
run_series
```

```python
#custom index
marks=[98,92,87,81,79]
subjects=['Math','HinduDharam','Chemistry','Biology','Physics']
pd.Series(marks,index=subjects)
```

```python
#Settings a name
marks=pd.Series(marks,index=subjects,name='Hridoy ke marks')
marks
```

###Series From Dict

```python
marks={'math':98,
       'HinduDharam':92,
       'Chemistry':87,
       'Biology':81,
       'Physics':79
       }

marks_series=pd.Series(marks,name='Hridoy ke marks')
marks_series
```
